```python
import matplotlib.pyplot as plt
import numpy as np
import math
vinf=1000 #m/s mean velocity of molecular beam
vs= 50 # m/s deviation from central velocity 5%
dslit= 1e-3 # meter size of the slit
#r= 50e-3 # meter radius of the chopper wheel

S0= 20000 
S1= 40000 
S2= 60000 
S3= 80000
# RPS revolution per second of the chopper wheel
d = 50e-3 # Distance between the chopper wheel and the laser ionization volume

def cm(value): 
    return value/2.54
#w0=(vinf*dslit)/(2*3.24*r*S)
#ws=vs*d/vinf
beampacket=[]
beampacket1=[]
beampacket2=[]
beampacket3=[]
f2=[]
for r in np.arange (1e-3, 300e-3, 0.001):
    w0=(vinf*dslit)/(2*3.24*r*(S0/60))
    w01=(vinf*dslit)/(2*3.24*r*(S1/60))
    w02=(vinf*dslit)/(2*3.24*r*(S2/60))
    w03=(vinf*dslit)/(2*3.24*r*(S3/60))
    ws=vs*d/vinf
    W=np.sqrt((w0*w0)+(ws*ws))
    W1=np.sqrt((w01*w01)+(ws*ws))
    W2=np.sqrt((w02*w02)+(ws*ws))
    W3=np.sqrt((w03*w03)+(ws*ws))
    f2.append(2*r*10)
    #f2.append(r)
    beampacket.append((W/vinf)*1e6) ## pulsewidth is in microseconds
    beampacket1.append((W1/vinf)*1e6) 
    beampacket2.append((W2/vinf)*1e6) 
    beampacket3.append((W3/vinf)*1e6) 
fig02, ax = plt.subplots(1, 1, figsize=(cm(45),cm(25)))
ax.plot(f2,beampacket, lw=4,label='20000 RPM')
ax.plot(f2,beampacket1, lw=4,label='40000 RPM')
ax.plot(f2,beampacket2, lw=4,label='60000 RPM')
ax.plot(f2,beampacket3, lw=4,label='80000 RPM')
ax.set_ylim(0,20)
ax.tick_params(axis='both', which='major', labelsize=25)
ax.tick_params(axis='both', which='minor', labelsize=25)
ax.ticklabel_format(useOffset=False)
ax.grid()
#ax.legend()
ax.legend(fontsize=20, loc ="upper right")
ax.set_xlabel('Diameter of chopper wheel (cm)', fontsize=25)
ax.set_ylabel('Molecular beam pulse width ($\mu$s)', fontsize=25)
#fig02.savefig('Chopperdesigncalculation.pdf',bbox_inches='tight',dpi=500)
```
